```julia
using JuMP, GLPK, Plots

m = Model(GLPK.Optimizer)

#variable, objective, constraint

@variable(m, 0<= A)
@variable(m, 0<= B)

@objective(m, Max, 4A+B)

@constraint(m, constraint1, 10A + 2B <=30)
@constraint(m, constraint2, 3A + 2B <=12)
@constraint(m, constraint3, 2A + 2B <=10)

print(m)

JuMP.optimize!(m)

println("MAX=",objective_value(m))
println("A=",JuMP.value(A))
println("B=",JuMP.value(B))
            
A = 0:5
C1(A) = -5A +15 
C2(A) = -1.5A +6
C3(A) = -A +5
M(A) = -4A + 12.428

feasible = min(C1(A[end]),C2(A[end]),C3(A[end]))

plot(A,C1,fill = (feasible, 0.5, :auto),lw=2)
        
plot!(A,C2,fill = (feasible, 0.5, :auto),lw=2)

plot!(A,C3,fill = (feasible, 0.5, :auto),lw=2)
plot!(A,M, lc=:red, lw=3)
```
