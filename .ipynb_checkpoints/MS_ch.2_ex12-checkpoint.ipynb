```julia
using JuMP, GLPK

m = Model(GLPK.Optimizer)

# variable, objective, constraint
@variable(m, 0 <= A)
@variable(m, 0 <= B)

@objective(m, Max, 3A+3B)

@constraint(m, constraint1, 2A+4B<=12)
@constraint(m, constraint2, 6A+4B<=24)

print(m)

JuMP.optimize!(m)

println("Max = ", objective_value(m))
println("A = ", JuMP.value(A))
println("B = ", JuMP.value(B))    

using Plots
A = 0:5
C1(A) = 3-0.5*A
C2(A) = 6-1.5*A
M(A) = -A + 13.5/3

the_max = min(C1(A[1]),C2(A[end]))
    
plot(A, C1, fill = (the_max, 0.5, :auto))
plot!(A, C2, fill = (the_max, 0.5, :auto))
plot!(A,M)
```
