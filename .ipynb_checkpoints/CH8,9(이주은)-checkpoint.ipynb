```julia
using Optim
using Plots
#using PyPlot
using JuMP
using Ipopt
using LinearAlgebra
using Distributions
```

```julia
#########   ch.8     ############


function f_line(x_bar, d_bar, lambda)
    x_new = x_bar +lambda*d_bar
    
    return (x_new[1])^4 + 2(x_new[1])^2 + 2(x_new[1]*x_new[2])+ 4(x_new[2])^4
end
```

```julia
x_bar = [2.0 , 3.0]
d_bar = [-1.0 , 0.0 ]
```

```julia
opt = Optim.optimize(lambda ->f_line(x_bar, d_bar, lambda),-100.0 , 100.0 , GoldenSection() )
```

```julia
println("optimal lambda = ", Optim.minimizer(opt))
println("optimal f(x+lambda*d) = ",Optim.minimum(opt))
```
